## Investigation Overview

My presentation will be focusing on what time of the day and day of week more rides are taking putting into consideration the ride duration and how user_type and member gender affects ride time and duration.


## Dataset Overview

There are total of 183,412 bike rides record with 16 features. 197 of the rides records were dropped due to missing start and end station information. 8,265 of the records are missing the riders member_birth_year and member_gender information. The missing member gender was replaced with Not Applicable(NA). Extra features (duration_min, hour_of_the_day and start_day) were created to ease plotting. All rides with more than an hour duration were considered outliers for this presentation and were removed 


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
gobike = pd.read_csv('201902-fordgobike-tripdata.csv', parse_dates=['start_time', 'end_time'])

# fill miss member_gender
gobike["member_gender"].fillna('NA', inplace=True)

# duration in minutes
gobike['duration_min'] = gobike['duration_sec'] / 60

#add start_day column to dataset
gobike['start_day'] = gobike.start_time.dt.day_of_week

# extracting time as a new feature in float between 0 and 24 to see hour trips are distributed during the day.
gobike['hour_of_the_day'] = gobike.start_time.dt.hour + gobike.start_time.dt.minute/60

# missing_station_filter
missing_stations_filter =  gobike['start_station_id'].isna() | gobike['end_station_id'].isna()

# above_1hr_filter
above_1hr_filter = gobike["duration_min"] > 60

gobike = gobike[~(missing_stations_filter | above_1hr_filter)]
```

## Rides count per day of the week

More rides are taken during the week days (Monday to Firday), Thursday having the most rides. Weekend rides are just about half of the rides that happens on week days.

```python
base_color = sb.color_palette()[0]
day_of_week  = ['Monday', 'Tuesday', 'Wednessday', 'Thursday', "Friday", "Saturday", "Sunday"]
sb.countplot(data=gobike, x=gobike.start_time.dt.day_of_week, color=base_color)
plt.xticks(np.arange(0,7), day_of_week, rotation='45')
plt.xlabel('Day of the week')
plt.ylabel('Number of rides')
plt.title('Rides per day of week')

day_of_week_counts = gobike.start_time.dt.day_of_week.value_counts()
total_counts = day_of_week_counts.sum()

#print the proportion text on the bars
for i in range (day_of_week_counts.shape[0]):
    
    count = day_of_week_counts.loc[i]
    # Convert count into a percentage, and then into string
    pct_string = '{:0.1f}%'.format(100*count/total_counts)
    # Print the string value on the bar. 
    plt.text(i, count, pct_string, ha='center')
plt.show()
```

## Distribution of rides by time of the day
We notice that we have two peak time first one between 8am and 9am, also between 5pm and 6pm. There is also a slight increase in rides taken between midday and 1:30pm.

```python
bins = np.arange(0, gobike['hour_of_the_day'].max()+0.5, 0.5)
def time_of_hist(data, bins):
    plt.figure(figsize=[10, 5])
    plt.hist(data = data, x = 'hour_of_the_day', bins = bins)
    xticks_time = ["Midnight","1AM","2AM","3AM","4AM","5AM","6AM","7AM","8AM","9AM","10AM","11AM","Midday","1PM","2PM","3PM","4PM","5PM","6PM","7PM","8PM","9PM","10PM","11PM","Midnight"]
    plt.xticks(range(0,25), xticks_time, rotation=90)
    plt.xlabel("Time of the day")
    plt.ylabel("Number of rides")
    plt.title("Distribution of rides accross time to the day")
    plt.show()

time_of_hist(gobike, bins)
```

## Rides count by Gender

The male Gender took 71.4% of all rides in the dataset while the Female gender took 22.3%. 

```python
member_gender_counts = gobike["member_gender"].value_counts()

sb.countplot(data=gobike, x='member_gender', color=base_color, order=member_gender_counts.index)
# plt.xticks(np.arange(0,7), day_of_week, rotation='45')
plt.xlabel('Member Gender')
plt.ylabel('Number of rides')
plt.title('Total rides taking by each gender')


total_gender_counts = member_gender_counts.sum()

# #print the proportion text on the bars
for i in range (member_gender_counts.shape[0]):
    
    count = member_gender_counts[i]
    # Convert count into a percentage, and then into string
    pct_string = '{:0.1f}%'.format(100*count/total_gender_counts)
    # Print the string value on the bar. 
    plt.text(i, count, pct_string, ha='center')
# plt.show()
```

## Gender vs Day of week

For weekdays, the male percentage was around 71-72% but drops below 70% for saturdays and sundays. other gender have slight increase in percentage for weekends.

## Ride Duration vs Day of week

We will notice that, the week days Monday - Friday, roughly have same average ride duration of 10mins and same range while the weekends have higher average duration (almost 12mins) and range. For both the weekdays and weekends, there are outliers, starting at 25mins for weekdays and roughly 30mins for weekends.

```python
plt.figure(figsize = [20, 5])
base_color = sb.color_palette()[0]

# left plot: violin plot
plt.subplot(1, 3, 1)
sb.violinplot(data=gobike, x='start_day', y='duration_min', inner = 'box',
              color = base_color)
plt.xticks(np.arange(0,7), day_of_week, rotation='45')
plt.title("Weekday average ride duration in minutes")
plt.ylabel("Averag Duration (mins)")

# center plot: box plot
plt.subplot(1, 3, 2)
sb.boxplot(data=gobike, x='start_day', y='duration_min',  color = base_color)
plt.xticks(np.arange(0,7), day_of_week, rotation='45')
plt.ylabel("Averag Duration (mins)")
plt.title("Weekday average ride duration in minutes")

# right plot: adapted bar chart
plt.subplot(1, 3, 3)
sb.barplot(data=gobike, x='start_day', y='duration_min',  color = base_color, ci=None)
plt.xticks(np.arange(0,7), day_of_week, rotation='45')
plt.ylabel("Averag Duration (mins)")
plt.title("Weekday average ride duration in minutes")

plt.show()
```

## User type vs ride duration for each day of the week

The subscriber user type have rides for 10mins on average while the Customer have higher average time and ride longer during the weekends. The majority of the trips are done by the subscribers.

```python
g = sb.FacetGrid(data = gobike, col = 'user_type', height = 4)
g.map(sb.barplot, 'start_day', 'duration_min', order=range(0,7))
g.set_titles(col_template="{col_name} weekday average duration (mins)", row_template="{row_name} weekday average duration (mins)")
g.set_axis_labels("Start day", "Duration (mins)")
plt.xticks(np.arange(0,7), day_of_week, rotation='45')
for axes in g.axes.flat:
    _ = axes.set_xticklabels(axes.get_xticklabels(), rotation=45)

plt.show()


g = sb.FacetGrid(data = gobike, col = 'user_type', height = 4)
g.map(sb.boxplot, 'start_day', 'duration_min', order=range(0,7))
g.set_axis_labels("Start day", "Duration (mins)")
g.set_titles(col_template="{col_name} weekday average duration (mins)", row_template="{row_name} weekday average duration (mins)")
plt.xticks(np.arange(0,7), day_of_week, rotation='45')
for axes in g.axes.flat:
    _ = axes.set_xticklabels(axes.get_xticklabels(), rotation=45)
plt.show()
```
