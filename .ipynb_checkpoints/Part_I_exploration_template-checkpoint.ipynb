## Preliminary Wrangling


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe, print statistics
gobike = pd.read_csv('201902-fordgobike-tripdata.csv', parse_dates=['start_time', 'end_time'])
```

```python
# high-level overview of data shape and composition
print(gobike.shape)
print(gobike.dtypes)
print(gobike.head(10))
```

```python
# check missing values
gobike.isna().sum()
```

```python
# filter records with missing station inf0
missing_stations_filter =  gobike['start_station_id'].isna() | gobike['end_station_id'].isna()
gobike = gobike[~missing_stations_filter]
gobike
```

```python
# fill missing gender with NA
gobike["member_gender"].fillna('NA', inplace=True)
```

```python
#add start_day column to dataset
gobike['start_day'] = gobike.start_time.dt.day_of_week
```

```python
gobike.describe()
```

## Univariate Exploration

#### How are rides distributed accross day of the weeks

```python
base_color = sb.color_palette()[0]
day_of_week  = ['Monday', 'Tuesday', 'Wednessday', 'Thursday', "Friday", "Saturday", "Sunday"]
sb.countplot(data=gobike, x=gobike.start_time.dt.day_of_week, color=base_color)
plt.xticks(np.arange(0,7), day_of_week, rotation='45')
plt.xlabel('Day of the week')
plt.title('Rides per day of week')


day_of_week_counts = gobike.start_time.dt.day_of_week.value_counts()
total_counts = day_of_week_counts.sum()

#print the proportion text on the bars
for i in range (day_of_week_counts.shape[0]):
    
    count = day_of_week_counts.loc[i]
    # Convert count into a percentage, and then into string
    pct_string = '{:0.1f}%'.format(100*count/total_counts)
    # Print the string value on the bar. 
    plt.text(i, count, pct_string, ha='center')
plt.show()
```

Our assumption of having more rides during the weekdays is clearly shown in the above chart, we can now take a look at the overall distribution of rides accross hours of the day

We will start with a bin size equivalent to every 30mins from Midnight

```python
# extracting time as a new feature in float between 0 and 24 to see hour trips are distributed during the day.
gobike['hour_of_the_day'] = gobike.start_time.dt.hour + gobike.start_time.dt.minute/60
bins = np.arange(0, gobike['hour_of_the_day'].max()+0.5, 0.5)
def time_of_hist(data, bins):
    plt.figure(figsize=[10, 5])
    plt.hist(data = data, x = 'hour_of_the_day', bins = bins)
    xticks_time = ["Midnight","1AM","2AM","3AM","4AM","5AM","6AM","7AM","8AM","9AM","10AM","11AM","Midday","1PM","2PM","3PM","4PM","5PM","6PM","7PM","8PM","9PM","10PM","11PM","Midnight"]
    plt.xticks(range(0,25), xticks_time, rotation=90)
    plt.xlabel("Time of the day")
    plt.ylabel("Rides counts")
    plt.title("Distribution of rides accross time to the day")
    plt.show()

time_of_hist(gobike, bins)
```

lets' see an hourly bin size equivalent

```python
bins = np.arange(0, gobike['hour_of_the_day'].max()+0.5, 1)
time_of_hist(gobike, bins)
```

From above we notice we have two peak time first one between 8am and 9am, also between 5pm and 6pm. There is also a slight increase in rides taken between midday and 1:30pm. Next we will look at the distribution of ride time.

```python
# let's create addtional features for duration in minutes 
gobike['duration_min'] = gobike['duration_sec'] / 60

# All records using a binsize 60 - a minute
binsize = 60
bins = np.arange(0, gobike['duration_min'].max()+binsize, binsize)
plt.hist(data=gobike, x='duration_min', bins=bins)
plt.show()
```

#### We see that the rides look right skewed and have some significantly very high ride time. Let us separate rides that take longer than an hour and plot a new histogram

```python
# we see that the rides look ride skewed and have some significantly very high ride time
above_1hr_filter = gobike["duration_min"] > 60

gobike.loc[above_1hr_filter, 'above_1hr'] = 'Yes'
gobike.loc[~above_1hr_filter, 'above_1hr'] = 'No'
above_1hr_counts = gobike['above_1hr'].value_counts()

print(above_1hr_counts)
plt.pie(above_1hr_counts, labels = above_1hr_counts.index, counterclock = False);
plt.title("Ride duration Above 1hr")
plt.show()
```

The above shows that the data contains some outliers that should be separated for some of our findings. We will reproduced our histogram to see how the rides under 1hour are distributed

```python
# All records using a binsize 60 - a minute
gobike = gobike[gobike['duration_min'] <= 60]

binsize = 1
bins = np.arange(0, gobike['duration_min'].max()+binsize, binsize)
plt.hist(data = gobike, x = 'duration_min', bins = bins);
plt.xlabel("Duration in mins")
plt.show()
```

```python
# investigating using a binsize 2 minutes

binsize = 2
bins = np.arange(0, gobike['duration_min'].max()+binsize, binsize)
plt.hist(data = gobike, x = 'duration_min', bins = bins);
plt.show()
```

We can further investigate to see if transforming the histogram give further insights.

```python
# investigating using a binsize 2 minutes

log_binsize = 0.1
bins = 10 ** np.arange(0, np.log10(gobike['duration_min'].max())+log_binsize, log_binsize)
plt.hist(data = gobike, x = 'duration_min', bins = bins);
plt.xscale('log')
plt.xticks([1,10,100], [0,1,2])
plt.show()
```

The log transformation shows that this is a Log-normal distribution. 

Before we close this section, it will be good to see how the dataset is divided among genders

```python
# Where gender are missing, we will represent it as NotApplicable - NA
member_gender_counts = gobike["member_gender"].value_counts()

sb.countplot(data=gobike, x='member_gender', color=base_color, order=member_gender_counts.index)
# plt.xticks(np.arange(0,7), day_of_week, rotation='45')
plt.xlabel('Member Gender')
plt.ylabel('Number of rides')
plt.title('Total rides taking by each gender')


total_gender_counts = member_gender_counts.sum()

# #print the proportion text on the bars
for i in range (member_gender_counts.shape[0]):
    
    count = member_gender_counts[i]
    # Convert count into a percentage, and then into string
    pct_string = '{:0.1f}%'.format(100*count/total_gender_counts)
    # Print the string value on the bar. 
    plt.text(i, count, pct_string, ha='center')
# plt.show()
```

The Male gender rides alot compare to others, it will be good to further explore maybe the percentage participation have any significant changes between hour of the day, weekdays and weekends.

## Bivariate Exploration

To begin, let's see how the day of the week is compared with duration in minutes

```python
plt.figure(figsize = [20, 5])
base_color = sb.color_palette()[0]

# left plot: violin plot
plt.subplot(1, 3, 1)
sb.violinplot(data=gobike, x='start_day', y='duration_min', inner = 'box',
              color = base_color)
plt.xticks(np.arange(0,7), day_of_week, rotation='45')


# center plot: box plot
plt.subplot(1, 3, 2)
sb.boxplot(data=gobike, x='start_day', y='duration_min',  color = base_color)
plt.xticks(np.arange(0,7), day_of_week, rotation='45')


# right plot: adapted bar chart
plt.subplot(1, 3, 3)
sb.barplot(data=gobike, x='start_day', y='duration_min',  color = base_color, ci=None)
plt.xticks(np.arange(0,7), day_of_week, rotation='45')
plt.show()
```

We will notice that, the week days Monday - Friday, roughly have same average ride duration of 10mins and same range while the weekends have higher average duration (almost 12mins) and range. For both the weekdays and weekends, there are outliers, starting at 25mins for weekdays and roughly 30mins for weekends.


Next we will look at how rides are distributed accross the hour of the day by gender, to see if there are patterns changes.

```python
 bin_edges = np.arange(0, gobike['hour_of_the_day'].max()+0.5, 0.5)
g = sb.FacetGrid(data = gobike, hue = 'member_gender', height = 4, aspect=1.65)
g.map(plt.hist, "hour_of_the_day", bins = bin_edges, alpha=.4)
g.add_legend()

xticks_time = ["Midnight","1AM","2AM","3AM","4AM","5AM","6AM","7AM","8AM","9AM","10AM","11AM","Midday","1PM","2PM","3PM","4PM","5PM","6PM","7PM","8PM","9PM","10PM","11PM","Midnight"]
plt.xticks(range(0,25), xticks_time, rotation=90)
plt.xlabel("Time of the day")
plt.ylabel("Rides counts")
plt.title("Distribution of rides accross time to the day")
plt.show()
```

```python
 
bin_edges = np.arange(0, gobike['hour_of_the_day'].max()+0.5, 0.5)
g = sb.FacetGrid(data = gobike, hue = 'member_gender', col="start_day", col_wrap=3 ,height = 4, aspect=1.65)
g.map(plt.hist, "hour_of_the_day", bins = bin_edges, alpha=.4)
g.add_legend()

xticks_time = ["Midnight","1AM","2AM","3AM","4AM","5AM","6AM","7AM","8AM","9AM","10AM","11AM","Midday","1PM","2PM","3PM","4PM","5PM","6PM","7PM","8PM","9PM","10PM","11PM","Midnight"]

for axes in g.axes.flat:
    _ = axes.set_xticklabels(axes.get_xticklabels(), rotation=90)
    
plt.xticks(range(0,25), xticks_time, rotation=90)
plt.xlabel("Time of the day")
plt.ylabel("Rides counts")
plt.title("Distribution of rides accross time to the day")
plt.show()
```

From the above we see that for monday, wednesday and thursday, there are closely same number of rides taking at the morning and evening peak while for Tuesday and Friday day, there is a significant difference between the morning peak and evening peaks. Midday ride peak varies for each day of Monday to friday, tuesday being close to a uniform distribution at midday day. Weekends differs totally showing a normal distribution starting from 7AM to Midgnight for Saturday and sundays

```python
plt.figure(figsize = [16, 4])

ax = sb.countplot(data = gobike, x = 'start_day', hue = 'member_gender')

totals = gobike.groupby('start_day')['start_time'].count()
plt.xticks(np.arange(0,7), day_of_week, rotation='0')

# create tabular form for bar heights
bars_height = []
for i in range(4):
    ind = i*7
    p_heights = [p.get_height() for p in ax.patches[ind:ind+7]]
    bars_height.append(p_heights)

# use pandas to get % from tabular form for bar heights
height_df = pd.DataFrame(bars_height)
height_df_percent = (100. * height_df / height_df.sum()).round(1).astype(str) + '%'

for i in range(4):    
    ind = i*7
    for p_ind, p in enumerate(ax.patches[ind:ind+7]):
        x = p.get_x() + 0.12
        y = p.get_height()
        ax.annotate(height_df_percent.loc[i][p_ind], (x, y), ha='center', va='bottom')
```

For weekdays, the male percentage was around 71-72% but drops below 70% for saturdays and sundays. other gender have slight increase in percentage for weekends.
